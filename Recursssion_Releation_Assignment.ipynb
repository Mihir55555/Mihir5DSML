```python
1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.

Ans:

T(n) = 3T(n-1) + 12n , T(0)=5

Step 1:

T(1) = 3 * T(0) + 12 * 1
     = 3 * 5 + 12
     = 15 + 12 
     = 27
     
Step 2:

T(2) = 3 * T(1) + 12 * 2
     = 3 * 27 + 24
     = 81 + 24
     = 105
     
T(2) = 105
```
